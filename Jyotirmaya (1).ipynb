Import the libraries which we will go throughout the programme

```python
import numpy as np
import pandas as pd
import sklearn
```

Data that we are using to brain our model 
We are using boston house pricing dataset.

```python
from sklearn.datasets import load_boston
df = load_boston()
```

```python
df.keys()
```

```python
print(df.DESCR)
```

Convert the dataset into the pandas dataframe to make it easier to analyse data

```python
boston = pd.DataFrame(df.data, columns=df.feature_names)
boston.head(20)
```

Adding a new column of target value to the dataframe.

```python
boston['MEDV'] = df.target
boston.head()
```

To check if the dataset contains any null value or not

```python
boston.isnull()
```

isnull return True or False for each of the cell in the dataframe, but we can't go exploring all the cells to look for True values if any,so we use sum() function to count all the cells withs True value(i.e. Null cells)

```python
boston.isnull().sum()
```

We never train  on all the data that we have,we always make sure to atleast have a test dataset,which is different from the training dataset.

```python
from sklearn.model_selection import train_test_split
X = boston.drop('MEDV', axis=1)
Y = boston['MEDV']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.15, random_state = 5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

On testing set

```python
#Fitting model on the traning dataset
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train,y_train_predict)))

print("The model prference for training set ")
print('RMSE is {}'.format(rmse))
print("\n")
```

```python
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test,y_test_predict)))
print("The model preference  for the testing set")
print('RMSE is {}'.format(rmse))
```
